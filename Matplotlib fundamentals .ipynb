# Data visualization basics using Matplotlib

```python
# Import numpy for numerical computaion
# Import matplotlip for visualization

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.arange(1,10)
Y=np.arange(1,10)
```

```python
#Scatter plot
plt.scatter(X,Y,c='r')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter plot between X and Y')
plt.show()
```

```python
Y=X*X
```

```python
#plt plot
plt.plot(X,Y,'r*--',linewidth=2,markersize=4)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Continuous plot between X and Y')
plt.show()
```

```python
#Plotting Subplot
plt.subplot(2,2,1)#will divide in 2 by 2 matrix and 1 will denote position
plt.plot(X,Y,'r+')
plt.subplot(2,2,2)
plt.plot(X,Y,'b+')
plt.subplot(2,2,3)
plt.plot(X,Y,'g+')
plt.subplot(2,2,4)
plt.plot(X,Y,'r')
```

```python
#plotting an expression
x=np.arange(1,11)
y=3*x*x*x+15
plt.plot(x,y,'r*--')
plt.title('Plotting expression')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
#plotting sine function
#generating x and y data points using numpy
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
#ploting using plt
plt.plot(x,y)
plt.title('Sine function plot')
plt.xlabel('x-->')
plt.ylabel('Sinx-->')
plt.show()

```

```python
# plotting Sine and cosine function in Subplot
# Generating data to plot
x=np.arange(0,6*np.pi,0.1)
X_Sine=np.sin(x)
Y_cos=np.cos(x)
#plotting Sine nd cos function in subplot
plt.subplot(2,2,1)
plt.plot(x,X_Sine,'g*')
plt.subplot(2,2,3)
plt.plot(x,Y_cos,'r*')
plt.show()
```

```python
# Bar plot
#Creating two different set of  (x,y)

x1=np.array([10,24,14,16])
y1=np.array([6,16,14,18])
x2=np.array([8,20,12,14])
y2=np.array([8,18,16,20])
#plotting the bar plot#bar plot is just plotting the cordinates as bar. It is useful in comaparison of discrete dataset
plt.bar(x1,y1,align='center')
plt.bar(x2,y2,color='g')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Plot')
plt.show()
```

```python
#plotting histogram
#creating dataset to plot
a=np.array([12,6, 14, 16, 20, 40, 60, 70,22,18,8,30,45,35])
#plotting histogram-Histogram plots density or count in range between bin size--by default binsize=10 we can 
#increase or decrease
plt.hist(a)
plt.title('Histogram')
```

```python
#plotting box plot
#creating data to plot
a=[np.random.normal(0,std,100) for std in np.arange(1,5)]
#a
#plotting box plot
plt.boxplot(a,vert=True,patch_artist=True)#vert=False to plot box plot horizontally
plt.title('Box Plot')
plt.show()
```

```python
#plotting pie-chart
#creating data to plot pie chart
labels=['Sports', 'Gym', 'Running', 'Yoga']
people=[400, 300, 200, 100]
color=['green', 'red', 'lightcoral', 'lightskyblue']
explode=[0.2,0,0,0]
#plotting pie chart
plt.pie(people,explode=explode,labels=labels,colors=color,autopct='%1.1f%%',shadow=True)
plt.title('Plotting prefrence of people')
plt.axis('equal')
plt.show()
```
